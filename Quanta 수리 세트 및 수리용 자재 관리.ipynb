### Quanta 불량 현황 추출

```python
물류불량 = pd.read_excel(fn, sheet_name=0)
물류불량 = 물류불량[['모델', 'S/N', '비고', 'Remark']]
물류불량.columns = ['Mapping Model.Suffix', 'S/N', 'Defective_Symptom(KOR)', 'Remark']
물류불량 = 물류불량[물류불량['Mapping Model.Suffix'].notnull()]
물류불량['Mapping Model.Suffix'] = 물류불량['Mapping Model.Suffix'].str.strip()
물류불량['S/N'] = 물류불량['S/N'].str.strip()
물류불량.insert(0, 'Vendor', 물류불량['Mapping Model.Suffix'].apply(lambda x:x.split('-')[0]).replace(srt_model).replace(vendor_find))
물류불량 = 물류불량[물류불량['Vendor'] == 'Quanta']
물류불량['Defective_Symptom(KOR)'] = 물류불량['Defective_Symptom(KOR)'].str.replace('\n', ',', regex=False)
물류불량 = 물류불량.reset_index(drop=True)
물류불량
```

```python
물류불량[~물류불량['Mapping Model.Suffix'].apply(check_model_name)] # 모델명 오류가 없는지 확인
```

```python
물류불량[물류불량['S/N'].str.len() != 13] # S/N 자리수가 13자리가 아닌 것 확인
```

```python
quanta_doa_list = 물류불량[물류불량['Remark']=='DOA'].reset_index(drop=True)
quanta_doa_list.insert(4, 'Defective_Symptom(ENG)', quanta_doa_list['Defective_Symptom(KOR)'].apply(get_google_translate, args=['en']))
```

```python
quanta_doa_list[['Defective_Symptom(KOR)', 'Defective_Symptom(ENG)']].iloc[:30]
```

```python
# 번역 수정시
key_word = 'T/Pad Button 둔감'
# 수정할 데이터 확인
quanta_doa_list[quanta_doa_list['Defective_Symptom(KOR)'].str.contains(key_word)]
```

```python
# 수정
quanta_doa_list.loc[quanta_doa_list['Defective_Symptom(ENG)'].str.contains(key_word), 'Defective_Symptom(ENG)'] = 'T/Pad Step, T/Pad Button insensitivity'
```

```python
box_damage_list = 물류불량[물류불량['Remark'].str.contains('Box')].reset_index(drop=True)
box_damage_list['Defective_Symptom(KOR)'] = 'Box 불량'
box_damage_list.insert(4, 'Defective_Symptom(ENG)', 'Box damage')
box_damage_list.insert(5, 'Box P/N', None)
box_damage_list
```

```python
file_name = f"{datetime.datetime.now().strftime('%y%m%d')}_Quanta_DOA_List.xlsx" #이번주 forecast 파일을 저장할 이름 설정
with pd.ExcelWriter("D:/Python result/"+file_name) as writer:
    quanta_doa_list.to_excel(writer, sheet_name='DOA list', index=True)
    box_damage_list.to_excel(writer, sheet_name='Box dmamge list', index=True)
```

## 최종 작업한 수리리스트를 불러옴

```python
수리리스트3 = pd.read_excel(get_filename(), sheet_name=1)
```

```python
수리리스트3
```

```python
get_pdr().columns
```

```python
컬러테이블 = get_pdr()[['Model.Suffix', '1.2 SET COLOR']]
컬러테이블.columns = ['Mapping Model.Suffix', 'Color']
컬러테이블
```

```python
수리리스트3 = 수리리스트3.merge(컬러테이블)
수리리스트3
```

```python
수리리스트3['Repair Item'] = 수리리스트3['Repair Item'].str.split(',')
수리리스트3
```

```python
수리리스트3.to_clipboard()
```

```python
repair_items = pd.DataFrame()
i = 0
for row in 수리리스트3.iterrows():
    for item in row[1]['Repair Item']:
        repair_items.loc[i, ['Series', 'Mapping Model.Suffix', 'S/N', 'Color', 'Material Name']] = [
            row[1]['Series'],
            row[1]['Mapping Model.Suffix'],
            row[1]['S/N'],
            row[1]['Color'], item]
        i += 1
repair_items
```

```python
repair_items['Material Name'] = repair_items['Material Name'].str.strip().str.lower().str.capitalize()
repair_items = repair_items.fillna('-')
repair_items.groupby(['Series', 'Mapping Model.Suffix', 'S/N', 'Color', 'Material Name'])['Series'].count().unstack().convert_dtypes().fillna(0).to_clipboard()
```
