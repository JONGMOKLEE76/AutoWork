## 기존 Box 정보에 공급일정이나 결과 update

```python
updated_df = pd.read_clipboard() # 새로운 정보가 업데이트된 엑셀을 불러와 데이터프레임으로 변환
updated_df
```

```python
for row in updated_df.iterrows(): # 불러온 데이터프레임의 각 row 데이터를 불러와 box DB의 공급결과에 업데이트함
    c1 = (box['Request Date'] == row[1]['Request Date'])
    c2 = (box['To Site'] == row[1]['To Site'])
    c3 = (box['Mapping Model.Suffix'] == row[1]['Mapping Model.Suffix'])
    c4 = (box['QTY'] == row[1]['QTY'])
    box.loc[c1 & c2 & c3 & c4, ['B/L Number', 'Container Number']] = row[1]['B/L Number'], row[1]['Container Number']
```

```python
box # DB에 업데이트 되었는지 확인
```

```python
# DB 저장
with open('D:/Data/TCL box.bin', 'wb') as f:
    pickle.dump(box, f)
```
