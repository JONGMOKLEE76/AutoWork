## DPK Blanket PO 발행 완료 후

```python
with open('D:/Data/DPK blanket PO DB.bin', 'rb') as f:
    bpo = pickle.load(f)
```

```python
df1 = pd.read_excel('D:/key order inquiry.xlsx')
df1
```

```python
df1 = df1.rename(columns={'LG\nPO No.':'LG PO No.', 'LG\nPO Date':'LG PO Date', 'MS\nOrder No.':'MS Order No.', 'MS P/No.':'MS P/N', 'LG P/No.':'LG P/N'})
df1
```

```python
df1 = df1.drop(columns=['Site', 'Key Type', 'Fulfillment\nQty', 'Gap Qty', 'Order Status'])
df1['Order Qty'] = df1['Order Qty'].astype(str).apply(lambda x:int(x.replace(',', '')))
df1['LG PO Date'] = pd.to_datetime(df1['LG PO Date'])
df1
```

```python
bpo.tail(10)
```

```python
for row in df1.iterrows():
    c1 = (bpo['Request Qty'] == row[1]['Order Qty'])
    c2 = (bpo['LG P/N'] == row[1]['LG P/N'])
    c3 = (bpo['Order Type'].isnull())
    bpo.loc[c1 & c2 & c3, ['Order Type', 'LG PO No.', 'LG PO Date', 'Order Qty', 'MS Order No.']] = [
        row[1]['Order Type'], row[1]['LG PO No.'], row[1]['LG PO Date'], row[1]['Order Qty'], row[1]['MS Order No.']]
```
