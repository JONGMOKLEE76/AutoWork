```python
def login_glop(place):
    if place == 'office':
        pyautogui.click(444, 23) # 주소 표시줄 클릭
        pyautogui.sleep(0.3)
        pyautogui.typewrite('glopp.lge.com/index.jsp', interval=0.1)
        pyautogui.press('enter', presses=2, interval=1)
        pyautogui.sleep(1)
        while not pyautogui.pixelMatchesColor(868, 619, (255, 255, 255)):  # GLOPP 최초 로그인 후 초기 정보 로딩까지 대기
            pyautogui.sleep(1)
        pyautogui.moveTo(926, 592, duration=0.3)
    elif place == 'home':
        pyautogui.click(682, 40) # 주소 표시줄 클릭
        pyautogui.sleep(0.3)
        pyautogui.typewrite('glopp.lge.com/index.jsp', interval=0.1)
        pyautogui.press('enter', presses=2, interval=1)
        pyautogui.sleep(1)
        while not pyautogui.pixelMatchesColor(868, 619, (255, 255, 255)):  # GLOPP 최초 로그인 후 초기 정보 로딩까지 대기
            pyautogui.sleep(1)
        pyautogui.moveTo(926, 592, duration=0.3)
    elif place == 'laptop':
        pyautogui.click(333, 27) # 주소 표시줄 클릭
        pyautogui.sleep(0.3)
        pyautogui.typewrite('glopp.lge.com/index.jsp', interval=0.1)
        pyautogui.press('enter', presses=2, interval=1)
        pyautogui.sleep(1)
        while not pyautogui.pixelMatchesColor(183, 973, (255, 255, 255)):  # GLOPP 최초 로그인 후 초기 정보 로딩까지 대기
            pyautogui.sleep(1)
        pyautogui.moveTo(545, 604, duration=0.3)

def select_odm(place, vendor):
    if place == 'office':
        pyautogui.click(1613, 121, duration=0.2) # 업체 선택 dropdown 메뉴 클릭
        pyautogui.moveTo(1710, 160, duration=0.5) # scroll bar를 클릭
        if vendor == 'Pegatron PC':
            pyautogui.dragRel(0, 30, duration=0.3) # scroll bar를 내림
            pyautogui.click(1580, 189, duration=0.2) # Pegatron PC 선택
        elif vendor == 'Pegatron Thin Client':
            pyautogui.dragRel(0, 30, duration=0.3) # scroll bar를 내림
            pyautogui.click(1579, 203, duration=0.2) # Pegatron Thin Client 선택
        elif vendor == 'Quanta':
            pyautogui.dragRel(0, 40, duration=0.3) # scroll bar를 내림
            pyautogui.click(1585, 175, duration=0.2) # Quanta 선택
        elif vendor == 'Wingtech':
            pyautogui.dragRel(0, 60, duration=0.3) # scroll bar를 내림
            pyautogui.click(1591, 220, duration=0.2) # Wingtech 선택
        elif vendor == 'TCL MOKA':
            pyautogui.dragRel(0, 40, duration=0.3) # scroll bar를 내림
            pyautogui.click(1585, 194, duration=0.2) # TCL MOKA 선택
        elif vendor == 'TPV':
            pyautogui.dragRel(0, 45, duration=0.3) # scroll bar를 내림
            pyautogui.click(1596, 192, duration=0.2) # TPV 선택
        elif vendor == 'AUO':
            pyautogui.click(1596, 142, duration=0.2) # TPV 선택

    elif place == 'home':
        pyautogui.click(898, 120, duration=0.2) # 업체 선택 dropdown 메뉴 클릭
        pyautogui.moveTo(1070, 159, duration=0.2) # scroll bar를 클릭
        if vendor == 'Pegatron PC':
            pyautogui.dragRel(0, 30, duration=0.3) # scroll bar를 내림
            pyautogui.click(935, 183, duration=0.2) # Pegatron PC 선택
        elif vendor == 'Pegatron Thin Client':
            pyautogui.dragRel(0, 30, duration=0.3) # scroll bar를 내림
            pyautogui.click(940, 202, duration=0.2) # Pegatron Thin Client 선택
        elif vendor == 'Quanta':
            pyautogui.dragRel(0, 30, duration=0.3) # scroll bar를 내림
            pyautogui.click(946, 221, duration=0.2) # Quanta 선택
        elif vendor == 'Wingtech':
            pyautogui.dragRel(0, 60, duration=0.3) # scroll bar를 내림
            pyautogui.click(953, 220, duration=0.2) # Wingtech 선택
        elif vendor == 'TCL MOKA':
            pyautogui.dragRel(0, 60, duration=0.3) # scroll bar를 내림
            pyautogui.click(957, 141, duration=0.2) # TCL MOKA 선택
        elif vendor == 'TPV':
            pyautogui.dragRel(0, 60, duration=0.3) # scroll bar를 내림
            pyautogui.click(958, 162, duration=0.2) # TPV 선택
        elif vendor == 'AUO':
            pyautogui.click(948, 142, duration=0.2) # AUO 선택
            
    elif place == 'laptop':
        pyautogui.click(943, 115, duration=0.2) # 업체 선택 dropdown 메뉴 클릭
        pyautogui.moveTo(944, 159, duration=0.5) # scroll bar를 클릭
        if vendor == 'Pegatron PC':
            pyautogui.dragRel(0, 30, duration=0.3) # scroll bar를 내림
            pyautogui.click(831, 180, duration=0.2) # Pegatron PC 선택
        elif vendor == 'Pegatron Thin Client':
            pyautogui.dragRel(0, 30, duration=0.3) # scroll bar를 내림
            pyautogui.click(826, 200, duration=0.2) # Pegatron Thin Client 선택
        elif vendor == 'Quanta':
            pyautogui.dragRel(0, 40, duration=0.3) # scroll bar를 내림
            pyautogui.click(822, 174, duration=0.2) # Quanta 선택
        elif vendor == 'Wingtech':
            pyautogui.dragRel(0, 60, duration=0.3) # scroll bar를 내림
            pyautogui.click(828, 220, duration=0.2) # Wingtech 선택
        elif vendor == 'TCL MOKA':
            pyautogui.dragRel(0, 40, duration=0.3) # scroll bar를 내림
            pyautogui.click(828, 193, duration=0.2) # TCL MOKA 선택
        elif vendor == 'TPV':
            pyautogui.dragRel(0, 45, duration=0.3) # scroll bar를 내림
            pyautogui.click(832, 188, duration=0.2) # TPV 선택
        elif vendor == 'AUO':
            pyautogui.click(833, 141, duration=0.2) # AUO 선택
    pyautogui.sleep(1)
    while not pyautogui.pixelMatchesColor(899, 905, (255, 255, 255)):  # 업체 선택 후 조회될때까지
        pyautogui.sleep(1)

def go_shipping_and_download_excel(place):
    if place == 'office':
        pyautogui.moveTo(613, 82, duration=0.5) # Shipping & Invoicing 선택
        pyautogui.click(610, 121, duration=0.5) # Shipping & Invoicing 하위 메뉴 선택
        pyautogui.sleep(1)
        while not pyautogui.pixelMatchesColor(276, 1407, (255, 255, 255)):  # Shipping & Invoicing 메뉴로 전환될 때까지
            pyautogui.sleep(1)
        pyautogui.click(1884, 186, duration=0.2)
        pyautogui.press('enter')
        print('GLOP에서 파일 다운로드를 시작합니다.')
        while not pyautogui.pixelMatchesColor(1350, 58, (74, 74, 74)):  # 파일 다운로드 완료될 때까지
            pyautogui.sleep(1)
        filepath = 'D:/Downloads/' + [file for file in os.listdir('D:/Downloads/') if 'poshThru' in file][-1]
        print('파일 다운로드 완료!', filepath, end=',')
        return filepath
    elif place == 'home':
        pyautogui.moveTo(600, 82, duration=0.5) # Shipping & Invoicing 선택
        pyautogui.click(597, 118, duration=0.5) # Shipping & Invoicing 하위 메뉴 선택
        pyautogui.sleep(1)
        while not pyautogui.pixelMatchesColor(148, 1281, (255, 255, 255)):  # Shipping & Invoicing 메뉴로 전환될 때까지
            pyautogui.sleep(1)
        pyautogui.click(1239, 183, duration=0.2)
        pyautogui.press('enter')
        print('GLOP에서 파일 다운로드를 시작합니다.')
        while not pyautogui.pixelMatchesColor(757, 58, (74, 74, 74)):  # 파일 다운로드 완료될 때까지
            pyautogui.sleep(1)
        pyautogui.sleep(1)
        filepath = 'D:/Downloads/' + [file for file in os.listdir('D:/Downloads/') if 'poshThru' in file][-1]
        print('파일 다운로드 완료!', filepath, end=',')
        return filepath
    elif place == 'laptop':
        pyautogui.moveTo(594, 78, duration=0.5) # Shipping & Invoicing 선택
        pyautogui.click(598, 120, duration=0.2) # Shipping & Invoicing 하위 메뉴 선택
        pyautogui.sleep(1)
        while not pyautogui.pixelMatchesColor(204, 900, (255, 255, 255)):  # Shipping & Invoicing 메뉴로 전환될 때까지
            pyautogui.sleep(1)
        pyautogui.moveTo(405, 1026)
        pyautogui.dragRel(270, 0, duration=0.3)
        pyautogui.click(920, 184, duration=0.2)
        pyautogui.press('enter')
        print('GLOP에서 파일 다운로드를 시작합니다.')
        while not pyautogui.pixelMatchesColor(479, 60, (74, 74, 74)):  # 파일 다운로드 완료될 때까지
            pyautogui.sleep(1)
        pyautogui.sleep(2)
        filepath = 'D:/Downloads/' + [file for file in os.listdir('D:/Downloads/') if 'poshThru' in file][-1]
        print('파일 다운로드 완료!', filepath, end=',')
        return filepath

def processing_GLOP_data(filepath, vendor):
    df = pd.read_excel(filepath, sheet_name='DataSet').iloc[:, :48]
    df = df[['Model', 'PO No.', 'Ship To', 'Shipping', 'Cancel', 'PO', 'OQC Report', 'OQC Date', 'OQC Result', 'Ship', 'Issued Date', 'RSD','Ship Date', 'BL No', 'BL Status', 'Method', 'Price Term', 'Unit Price', 'Currency', 'Payment Term', 'PO1 No', 'SO No.', 'Final Destination']]
    df = df[~df['PO No.'].isin(['JGMZ23029072', 'JGMZ23027440', 'JGMZ23025674', 'JGMZ23029071'])] # cancel 되었으나 GLOP에서 'Y'로 되어 있어 강제로 제거 필요
    df.loc[df['PO1 No'].notnull(), 'PO1 No'] = df.loc[df['PO1 No'].notnull(),'PO1 No'].astype('int').astype('str')
    df.loc[df['SO No.'].notnull(), 'SO No.'] = df.loc[df['SO No.'].notnull(), 'SO No.'].astype('int').astype('str')
    df['Ship'].fillna(0, inplace=True)
    df['Ship'] = df['Ship'].astype('int')
    df['Issued Date'] = pd.to_datetime(df['Issued Date'])
    df['RSD'] = pd.to_datetime(df['RSD']) # PO의 RSD를 Datetime 형식으로 변경
    df.rename(columns={'Model':'Mapping Model.Suffix', 'Ship To':'To Site'}, inplace=True)
    div_map = {'Quanta':'PC', 'Pegatron Thin Client':'MNT', 'Pegatron PC':'PC', 'Wingtech':'PC', 'TCL MOKA':'MNT', 'TPV':'MNT', 'AUO':'MNT'}
    df['Division'] = div_map[vendor]
    df['Updated_at'] = datetime.datetime.now().strftime('%Y%m%d%H%M')
    return df

def update_PO_DB(dataframe, vendor):
    po = dataframe.groupby(['PO No.', 'Cancel', 'Issued Date', 'Mapping Model.Suffix', 'To Site', 'RSD', 'Method', 'Price Term', 'Unit Price', 'Currency', 'Payment Term', 'PO1 No', 'SO No.', 'Final Destination', 'Division', 'Updated_at'], dropna=False)[['PO']].sum().reset_index()
    po['From Site'] = 'OS_' + vendor.upper() + '_CN_P'
    po = po.set_index(['From Site', 'PO No.'])
    with open('D:/Data/ODM PO.db', 'rb') as f:
        po_db = pickle.load(f)
    original_po_db = po_db.copy()
    for i in po.index:
        po_db.loc[i, :] = po.loc[i, :]
    added_po = po_db.drop(index=original_po_db.index)
    with open('D:/Data/ODM PO.db', 'wb') as f:
        pickle.dump(po_db, f)
    with open('D:/Data/DB backup/ODM PO.db', 'wb') as f:
        pickle.dump(po_db, f)
    print(f"{vendor}의 PO DB를 업데이트 하였습니다. PO는 {po_db.shape[0] - original_po_db.shape[0]} 건 추가 되었습니다. ", end='')
    return added_po

def update_SR_DB(dataframe, vendor):
    with open('D:/Data/ODM_SR.db', 'rb') as f:
        SR_db = pickle.load(f)
    original_SR_db = SR_db.copy()
    sr = dataframe[dataframe['Ship Date'].notnull()]
    sr['Ship Date'] = pd.to_datetime(sr['Ship Date'])
    sr['From Site'] = 'OS_' + vendor.upper() + '_CN_P'
    sr = sr.set_index(['From Site', 'PO No.', 'BL No'])
    for i in sr.index:
        SR_db.loc[i, :] = sr.loc[i, :]
    added_sr = SR_db.drop(index=original_SR_db.index)
    with open('D:/Data/ODM PO.db', 'rb') as f:
        po_db = pickle.load(f)
    abnormal_open_po = pd.concat([po_db[po_db['Cancel'] == 'N']['PO'], SR_db.groupby(['From Site', 'PO No.']).sum('Ship')['Ship']], axis=1).fillna(0)
    abnormal_open_po['Open'] = abnormal_open_po['PO'] - abnormal_open_po['Ship']
    abnormal_open_po = abnormal_open_po[abnormal_open_po['Open'] < 0]
    if abnormal_open_po.shape[0] > 0:
        print('WARNING!! SR 업데이트하면서 Open PO 수량이 음수가 되었습니다. 선적 실적이 초과된 것이 없는 지 확인하세요!')
        sys.exit()
    else:
        with open('D:/Data/ODM_SR.db', 'wb') as f:
            pickle.dump(SR_db, f)
        with open('D:/Data/DB backup/ODM_SR.db', 'wb') as f:
            pickle.dump(SR_db, f)
        print(f"{vendor}의 SR DB를 업데이트 하였습니다. 선적은 {SR_db.shape[0] - original_SR_db.shape[0]} 건 추가 되었습니다. ")
        return added_sr
```

```python
vendor_list = ['AUO', 'Quanta', 'Pegatron Thin Client', 'Pegatron PC', 'Wingtech', 'TCL MOKA', 'TPV']
# vendor_list = ['TCL MOKA', 'TPV']
place = 'office'
added_po_total = pd.DataFrame()
added_sr_total = pd.DataFrame()
for num, vendor in enumerate(vendor_list):
    print(f'{num+1}_{vendor}의 GLOP 접속을 시작합니다.')
    login_glop(place)
    select_odm(place, vendor)
    filepath = go_shipping_and_download_excel(place)
    df = processing_GLOP_data(filepath, vendor)
    added_po = update_PO_DB(df, vendor.split(' ')[0])
    added_po_total = pd.concat([added_po_total, added_po])
    added_sr = update_SR_DB(df, vendor.split(' ')[0])
    added_sr_total = pd.concat([added_sr_total, added_sr])
```
