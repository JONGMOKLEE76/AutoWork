## 권혁호 주임이 추가로 전달한 불량 세트 리스트

```python
fname2 = get_filename() # 파일명을 선택
```

```python
df4 = pd.read_excel(fname2, sheet_name=0) # 파일을 불러옴
df4 = df4[['분류', '모델', 'S/N', '비고']] # 필요한 컬럼 선택
df4.columns  = ['Vendor', 'Mapping Model.Suffix', 'S/N', 'Defective_Symptom(KOR)'] # 컬럼명 재설정
df4 = df4.dropna(how='all') # NA 행 삭제
df4['Mapping Model.Suffix'] = df4['Mapping Model.Suffix'].str.strip() # 모델명 공백 정리
df4['S/N'] = df4['S/N'].str.strip() # S/N 공백 정리
df4['Defective_Symptom(KOR)'] = df4['Defective_Symptom(KOR)'].str.strip() # 불량명 공백 정리
df4['Defective_Symptom(KOR)'] = df4['Defective_Symptom(KOR)'].str.replace('\n', ',') # 개행문자(\n) 을 쉼표(,) 로 대체
df4.insert(0, 'Vendor_check', df4['Mapping Model.Suffix'].apply(lambda x:x.split('-')[0]).replace(srt_model).replace(vendor_find))
df4
```

```python
df4[~(df4['Vendor_check'] == df4['Vendor'])]['Vendor'].unique() # Vendor 명이 모델명으로 봤을 때 제대로 mapping되지 않은 것만 확인해 봄
```

```python
df4[~(df4['Vendor_check'] == df4['Vendor'])]['Vendor_check'].unique() # Vendor 명이 모델명으로 봤을 때 제대로 mapping되지 않은 것만 확인해 봄
```

```python
df4 = df4[df4['Vendor_check']==df4['Vendor']] # Quanta, Pegatron 불량만 sorting함
```

```python
df4.insert(1, 'Series', df4['Mapping Model.Suffix'].apply(lambda x:x.split('-')[0]).replace(srt_model)) # 시리즈명을 컬럼으로 추가
```

```python
df4 = df4.drop('Vendor_check', axis=1)
```
