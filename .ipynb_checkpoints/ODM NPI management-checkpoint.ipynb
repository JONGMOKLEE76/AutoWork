```python
df = pd.concat([mt_week.set_index(index_list).iloc[0],                                 # Material supply plan
                mt_week.set_index(index_list).iloc[0].shift(1, fill_value=0),          # Input Finished
#                 mt_week.set_index(index_list).iloc[0].shift(1, fill_value=0).cumsum(), # Input finished 누적
                qt_sp_by_color.loc[(model, color.upper())],                            # SP
#                 qt_sp_by_color.loc[(model, color.upper())].cumsum(),                   # SP 누적
#                 mt_week.set_index(index_list).iloc[0].shift(1, fill_value=0).cumsum() - qt_sp_by_color.loc[(model, color.upper())].cumsum()# Input Finished 누적 - SP 누적 차이
               ],  axis=1).fillna(0)
df.columns = ['Material', 'Input', 'SP']
df = df.sort_index()
df
```

```python
df.loc[df.index[0], ('Input(BOH)', 'CO')] = (df.loc[df.index[0], 'Input'], 0) # 첫번째 주차의 BOH 초기값은 별도로 입력해 줌(여기서는 기존 재고가 없고 해당 주차의 Input이 BOH인 것으로 가정함)
```

```python
df.loc[week, 'SP(E)'] = df.loc[week, 'SP'] + df.loc[week, 'CO']
```

```python
df.loc[get_weekname_from(week, 1), 'Input(BOH)'] = df.loc[week, 'Input(BOH)'] - df.loc[week, 'SP(E)']
```

```python
df.loc[get_weekname_from(week, 1), 'Input(BOH)'] = df.loc[week, 'Input(BOH)'] - df.loc[week, 'SP(E)']
```

```python
df.loc[get_weekname_from(week, 1), 'CO'] = 0
```

```python
df.head()
```

```python
week = df.index[1]
```

```python
df.loc[get_weekname_from(week, 1), 'Input(BOH)'] = df.loc[week, 'Input(BOH)'] - df.loc[week, 'SP(E)']
```

```python
for week in df.index:
    if df.loc[week, 'SP'] + df.loc[week, 'CO'] > df.loc[week, 'Input(BOH)']:
        df.loc[week, 'SP(E)'] = df.loc[week, 'Input(BOH)']
        df.loc[get_weekname_from(week, 1), 'Input(BOH)'] = df.loc[week, 'Input(BOH)'] - df.loc[week, 'SP(E)'] + df.loc[week, 'Input']
        df.loc[get_weekname_from(week, 1), 'CO'] = df.loc[week, 'SP'] + df.loc[week, 'CO'] - df.loc[week, 'SP(E)']
    else:
        df.loc[week, 'SP(E)'] = df.loc[week, 'SP'] + df.loc[week, 'CO']
        df.loc[get_weekname_from(week, 1), 'Input(BOH)'] = df.loc[week, 'Input(BOH)'] - df.loc[week, 'SP(E)'] + df.loc[week, 'Input']
        df.loc[get_weekname_from(week, 1), 'CO'] = 0        
```
